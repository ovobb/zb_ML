```python
from sklearn.datasets import fetch_olivetti_faces

faces_all = fetch_olivetti_faces()
print(faces_all.DESCR)
```

```python
K = 20
faces = faces_all.images[faces_all.target==K]
```

```python
faces
```

```python
import matplotlib.pyplot as plt

# 원본 10장
N = 2
M = 5
fig = plt.figure(figsize=(10,5))
plt.subplots_adjust(top=1, bottom=0, hspace=0, wspace=0.05)

for n in range(N*M):
    ax = fig.add_subplot(N, M, n+1)
    ax.imshow(faces[n], cmap=plt.cm.bone)
    ax.grid(False)
    ax.xaxis.set_ticks([])
    ax.yaxis.set_ticks([])

plt.tight_layout()
plt.show()
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)

X = faces_all.data[faces_all.target == K]
```

```python
X.shape
```

```python
# 위의 이미지는 64x64픽셀로 되어있었을 것이다. 그리고 그 그림이 10장 있는것
import numpy as np
np.sqrt(4096)
```

```python
W = pca.fit_transform(X)
W.shape
```

```python
X_inv = pca.inverse_transform(W)
```

```python
X_inv.shape
```

```python
# pca한 그림
fig = plt.figure(figsize=(10,5))
plt.subplots_adjust(top=1, bottom=0, hspace=0, wspace=0.05)

for n in range(N*M):
    ax = fig.add_subplot(N, M, n+1)
    ax.imshow(X_inv[n].reshape(64,64), cmap=plt.cm.bone)
    ax.grid(False)
    ax.xaxis.set_ticks([])
    ax.yaxis.set_ticks([])

plt.tight_layout()
plt.show()
```

- 디테일한 부분은 살짝 다르지만 전체적으로 비슷한 것을 알 수 있다
- 화질이 떨어졌다고 생각하면 됨

### 원점과 두개의 eigen face

```python
face_mean = pca.mean_.reshape(64,64)
```

```python
face_p1 = pca.components_[0].reshape(64,64)
face_p2 = pca.components_[1].reshape(64,64)
```

```python
plt.figure(figsize=(12,7))
plt.subplot(131)
plt.imshow(face_mean, cmap=plt.cm.bone)
plt.grid(False); plt.xticks([]); plt.yticks([]); plt.title('mean')
plt.subplot(132)
plt.imshow(face_p1, cmap=plt.cm.bone)
plt.grid(False); plt.xticks([]); plt.yticks([]); plt.title('face_p1')
plt.subplot(133)
plt.imshow(face_p2, cmap=plt.cm.bone)
plt.grid(False); plt.xticks([]); plt.yticks([]); plt.title('face_p2')
plt.show();
```

```python
# 가중치 선정
import numpy as np

N = 2
M = 5
W = np.linspace(-5, 10, N*M)
```

```python
# face_p1에 가중치를 곱한만큼 평균에 더해줌. 그리고 그 그림을 그려봄

fig = plt.figure(figsize=(10,5))
plt.subplots_adjust(top=1, bottom=0, hspace=0, wspace=0.05)

for n in range(N*M):
    ax = fig.add_subplot(N, M, n+1)
    ax.imshow(face_mean + W[n] * face_p1, cmap=plt.cm.bone)
    ax.grid(False)
    ax.xaxis.set_ticks([])
    ax.yaxis.set_ticks([])
    plt.title('W : ' + str(round(W[n])))

plt.tight_layout()
plt.show()
```

```python
# face_p2에 가중치를 곱한만큼 평균에 더해줌. 그리고 그 그림을 그려봄

fig = plt.figure(figsize=(10,5))
plt.subplots_adjust(top=1, bottom=0, hspace=0, wspace=0.05)

for n in range(N*M):
    ax = fig.add_subplot(N, M, n+1)
    ax.imshow(face_mean + W[n] * face_p2, cmap=plt.cm.bone)
    ax.grid(False)
    ax.xaxis.set_ticks([])
    ax.yaxis.set_ticks([])
    plt.title('W : ' + str(round(W[n])))

plt.tight_layout()
plt.show()
```

- 두 개의 성분을 다 표현해보기

```python
nx, ny = (5,5)
x = np.linspace(-5, 8, nx)
y = np.linspace(-5, 8, ny)

w1, w2 = np.meshgrid(x, y)
```

```python
w1.shape
```

```python
w1 = w1.reshape(-1,) # -1은 앞에것은 그대로 두라는 뜻
w2 = w2.reshape(-1,)
w1.shape
```

```python
fig = plt.figure(figsize=(10,5))
plt.subplots_adjust(top=1, bottom=0, hspace=0, wspace=0.05)

for n in range(N*M):
    ax = fig.add_subplot(N, M, n+1)
    ax.imshow(face_mean + w1[n] * face_p1 + w2[n] * face_p2, cmap=plt.cm.bone)
    ax.grid(False)
    ax.xaxis.set_ticks([])
    ax.yaxis.set_ticks([])
    plt.title(str(round(w1[n])) + ',' + str(round(w2[n])))

plt.tight_layout()
plt.show()
```
